### Game Animation

```python
from Animation.game_animation import make_plot
from bokeh.io import show, output_notebook
import pandas as pd
from functools import partial

output_notebook()
```

```python
# Define variables
x_range=(-52.5,52.5)
y_range=(-34, 34)

id_def = 2
id_att = 1

image_url = 'static/images/soccer.png'

# Read in data
df = pd.read_csv('sample_data/soccer_sample.csv', index_col=0)
```

The only thing you have to do to get this to work with your tracking data is to get it a pandas dataframe and make sure that it has columns for `x,y,team_id,player_id,time`.
The `x` and `y` coordinates are in the range `[-52.5, 52.5]`, `[-34, 34]`. This might chance for your data.

```python
df.head()
```

```python
make_anim_plot = partial(make_plot, df=df,image_url=image_url, id_def=id_def, id_att = id_att,
                           x_range=x_range, y_range=y_range, slider_steps=1,
                           headers = ["x", "y", "team_id", "player_id","time"], 
                           anim_speed=60,show_dist_speed=False)

show(make_anim_plot, notebook_url='localhost:8888')
```
